
### **List comprehensions**



```python
Lista1 = [];
for valor in range(5):            #Forma normal 
  Lista1.append(valor+10);
Lista1  
```

```python
Lista2 = [valor+10 for valor in range(5)];
Lista2
```

```python
Lista3 = [];
for numero in range(1,30): 
  if numero % 4 ==0: 
     Lista3.append(numero);           #Forma normal 
  
Lista3;
```

```python
multiplos4 = [ numero for numero in range(1,30) if numero % 4 == 0];
print(multiplos4);
```

```python
conceito = ['azul' if nota >=6 else 'vermelha' for nota in range(1,11)]
print(conceito);
```
