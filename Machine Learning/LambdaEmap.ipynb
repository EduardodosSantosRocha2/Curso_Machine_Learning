### ***Lambda e map***


## **Funcao normal**


```python
def area_quadrado(parametro):
  area = parametro**2;
  print(area); 
```

```python
area_quadrado(4);
```

Usando Funcao Lambda



```python
area_quadrado2 = lambda x: x**2;
```

```python
print(area_quadrado2(4));
```

```python
area_retangulo = lambda b,h: b*h;
```

```python
print(area_retangulo(4,7));
```

```python
L = [4,5,6,7,11,9,10];
areas = list(map(lambda x: x**2, L));
areas
```
