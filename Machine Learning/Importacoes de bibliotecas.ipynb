```python
from math import sqrt, factorial
n = 24;
raiz = math.sqrt(n);
print("A raiz de {} é {:.3f}".format(n,raiz));

fatorialde4 = math.factorial(4);
print("O fatorial de {} é {}".format(4,fatorialde4));
```
