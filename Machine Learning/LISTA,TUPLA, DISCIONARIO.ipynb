# ***LISTA [ ]***



```python
lista = [2,3,7,6]; #lista ou vetor
size = len(lista); #tamanho da lista

for i in lista:
  print(i);
lista.append(8); # inclui o valor 8 no final da lista
  
print("\n\nSegunda lista:")  
for i in lista:
  print(i);


lista.remove(8); # exclui o valor 8 no final da lista
  
print("\n\nlista ex:")  
for i in lista:
  print(i);  



lista.sort(); # ordem crescente
print("\n\lista em ordem crescente :")  
for i in lista:
  print(i);


lista.insert(1,2); # inclui o valor 2 na posicao escolhida da lista (no caso pos = 1)
  
print("\n\nSegunda lista:")  
for i in lista:
  print(i);


lista.pop(4); # exclui o valor 8 na posicao escolhida da lista (no caso pos = 4)
  
print("\n\nSegunda lista:")  
for i in lista:
  print(i);  

```

### ***DISCIONARIO { }***

```python
PESSOAS  = {'nome': "jose", "sexo": "M", "idade": 98}; # precisa de chave e valor
PESSOAS["time"] = "atletico-mg"; #adicionar uma chave e um valor
print(PESSOAS);

del PESSOAS["time"]; #deletar algo
print(PESSOAS);
```
